## Wykresy prezentujące rozkład zmiennych 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Freedom to make life choices'], bins=15, kde=True)
plt.title('variable distribution: freedom to make life choices')
plt.xlabel('value of freedom to make life choices')
plt.ylabel('frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['Healthy life expectancy at birth'], bins=20, kde=True)
plt.title('variable distribution: healthy life expectancy at birth')
plt.xlabel('value of healthy life expectancy at birth')
plt.ylabel('frequency')

plt.tight_layout()
plt.show()
```

**Wnioski z wykresów**

Wykresy obu rozkładów zmiennych kształtem przypominają rozkłady beta z odpowiednio dobranymi parametrami:

```python
from scipy.stats import beta

plt.figure(figsize=(15, 6))

a, b = 4, 2
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Beta Distribution Curve (a=4, b=2)')

a, b = 8, 2
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('Beta Distribution Curve (a=8, b=2)')

plt.tight_layout()
plt.show()
```

W przypadku rozkładu zmiennej: wolność dokonywania wyborów życiowych możemy zauważyć, że większość wyników jest w obszarze 0.7-0.9 co jest bardzo pozytywnym wynikiem. Wskazuje to na faktyczną możliwość ludzi decydowania o swojej przyszłości. Można też wyciągnąć wniosek, że prawdopodobnie znamienna wiekszość rekordów z wysokim czynnikiem pochodzi z państw wysoko rozwiniętych, gdzie ludzie mają faktyczną większą swobodę życia.

W przypadku rozkładu oczekiwanej długości życia podczas porodu, wartość z największą częstością występowania - około 67 lat jest bliska faktycznej średniej długości życia wynoszącej około 73 lata. Jest to spodziewany wynik, biorąc pod uwagę, że średnia jest zawyżona przez państwa bogate, w których wysoka jakość życia przeykłada się na długość życia.

## Wykresy wskazujące zależności w zbiorze danych

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='Freedom to make life choices', data=df, color='pink')
plt.xticks(rotation=45)
plt.title('Variation in Freedom to Make Life Choices Over Time')
plt.show()
```

**Wnioski**

Analizując wykres jak lekko odstający od reszty rzuca się okres lat 2005-2010, gdzie to obserwujemy spadek w wolność dokonywania wyborów życiowych. Można doszukać się korelacji ze szczytem globalnego kryzysu finansowego mającego miejsce w latach 2008-2009. Zmniejszona przez to swoboda finansowa mogła przełożyć się na mniejszą swobodę w dokonywaniu wyborów życiowych. 
Na wykresie widoczne są także wartości odstające (outliers), co może wskazywać na wyjątkowe przypadki krajów z bardzo niskim poziomem wolności.
Po 2010 roku można zauważyć wzrost mediany i jej stabilizację na wyższym poziomie, co sugeruje ogólną poprawę w odczuwaniu wolności do podejmowania decyzji życiowych.
Rozpiętość danych w późniejszych latach jest mniejsza niż w okresie 2005-2010, co wskazuje na bardziej jednolite poczucie wolności w różnych krajach.
Wartości odstające (punkty poniżej dolnego wąsa pudełka) są obecne w różnych latach, co wskazuje na istnienie krajów, w których poziom wolności do podejmowania decyzji życiowych jest znacznie niższy niż w większości innych krajów w danym roku.
W latach 2022-2023 zauważalna jest tendencja do dalszej stabilizacji lub niewielkiego wzrostu mediany, choć zakres danych nadal wskazuje na pewne zróżnicowanie.

```python
x = df['Life Ladder']
y = df['Log GDP per capita']
hue = df['Positive affect'] 

plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, hue=hue, palette='viridis')
plt.legend(title='Positive affect')
plt.title('Life Ladder vs. Log GDP per capita with Positive affect')
plt.xlabel('Life Ladder')
plt.ylabel('Log GDP per capita')

plt.show()
```

**Wnioski**

Na wykresie widać pozytywną korelację miedzy poziomem szczęścia (life ladder) a wartością PKB na człowieka. Wynika z tego, że w krajach o wyższym poziomie PKB per capita ludzie mają tendencję do wyższego poziomu zadowolenia z życia. Wyższy poziom pozytywnego afektu (jaśniejsze punkty) jest bardziej skoncentrowany w wyższych wartościach zarówno dla "Life Ladder", jak i PKB per capita. To sugeruje, że pozytywny afekt może być wyższy w krajach o wyższym zadowoleniu z życia i wyższym PKB per capita. Chociaż istnieje ogólna pozytywna korelacja, dane wykazują pewną dyspersję. Wartości "Life Ladder" dla danego poziomu PKB per capita mogą się różnić, co sugeruje, że inne czynniki również mają wpływ na zadowolenie z życia.
Podobnie, różne poziomy pozytywnego afektu występują przy różnych poziomach PKB per capita i "Life Ladder", co również wskazuje na złożoność wpływu różnych czynników na zadowolenie z życia.

Wykres pokazuje, że wyższe PKB per capita jest związane z wyższym poziomem zadowolenia z życia, a wyższy pozytywny afekt częściej występuje w krajach o wyższym zadowoleniu z życia i wyższym PKB per capita. Jednakże istnieje znaczna dyspersja danych, co sugeruje wpływ wielu czynników na te relacje.
