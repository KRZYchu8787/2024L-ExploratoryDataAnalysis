# Opis zbioru danych

Zbiór *energia* zawiera szczegółowe informacje dotyczące odnawialnych źródeł energii, w tym jej zużycia, przechowywania, produkcji, wpływu na środowisko oraz dane dt. inwestycji w energię odnawialną. Ramka posiada następujące kolumny:


1. Type_of_Renewable_Energy: Typ energii odnawialnej (1: Słoneczna, 2: Wiatrowa, 3: Turbiny wodne, 4: Geotermalna, 5: Biomasa, 6: Pływowa, 7: Fal).
2. Installed_Capacity_MW: Moc w megawatach (MW).
3. Energy_Production_MWh: Roczna produkcja energii w megawatogodzinach (MWh).
4. Energy_Consumption_MWh: Roczna konsumpcja energii w megawatogodzinach (MWh).
5. Energy_Storage_Capacity_MWh: Pojemność magazynów energii w megawatogodzinach (MWh).
6. Storage_Efficiency_Percentage: Efektywność systemów magazynowania energii w procentach.
7. Grid_Integration_Level: Poziom integracji sieci (1: W pełni zintegrowana, 2: Częściowo zintegrowana, 3: Zintegrowana w minimalnym stopniu, 4: Izolowana mikrosieć).
8. Initial_Investment_USD: Koszt wstępnej inwestycji w dolarach.
9. Funding_Sources: Źródło finansowania (1: Rząd, 2: Sektor Prywatny, 3: Mieszane).
10. Financial_Incentives_USD: Zachęty finansowe w dolarach.
11. GHG_Emission_Reduction_tCO2e: Redukcja emisji gazów cieplarnianych w odpowiedniku ton CO2 (tCO2e).
12. Air_Pollution_Reduction_Index: Redukcja indeksu zanieczyszczenia powietrza.
13. Jobs_Created: Liczba utworzonych miejsc pracy.

# Zależności między wybranymi zmiennymi

1. Finansowanie a ilość wytworzonych miejsc pracy

```python
energia[["Jobs_Created", "Funding_Sources"]].groupby("Funding_Sources").mean()
```

```python
energia[["Jobs_Created", "Funding_Sources"]].groupby("Funding_Sources").median()
```

Średnio więcej miejsc pracy generują inwestycje finansowane przez sektor prywatny

2. Typ energii a jej produkcja w MWh

```python
energia[["Energy_Production_MWh", "Type_of_Renewable_Energy"]].groupby("Type_of_Renewable_Energy").mean()
```

Średnio najwięcej MWh generuje energia z pływów

# Jakieś tam wykresy obrazujące rozkład

```python
sns.boxplot(data=energia,x='Funding_Sources',y='Jobs_Created')
```

Rozkład ilości utworzonych miejsc pracy jest bardzo podobny dla różnych źródeł finansowania

```python
plt.hist(energia[["Energy_Production_MWh"]], bins = 50)
```

Rozkład produkcji energii w MWh jest praktycznie jednostajny

# Jakieś tam wykresy obrazujące zależności

```python
energia[["Initial_Investment_USD", "Storage_Efficiency_Percentage"]].plot(x = "Initial_Investment_USD", y = "Storage_Efficiency_Percentage", kind = "scatter")
```

Nie ma istotnej zależności między wysokością wstępnej inwestycji a procentem efektywności przechowywania energii

```python
energia[["Initial_Investment_USD", "Type_of_Renewable_Energy"]].groupby("Type_of_Renewable_Energy").mean().plot(kind = "bar", logy = True)
```

Średnio największego kosztu wstępnego inwestycji wymagają panele fotowoltaiczne, a najmniejszego - energia pływowa.
