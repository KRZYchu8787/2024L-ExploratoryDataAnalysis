```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('energy_dataset_.csv')
```

```python
df.head()
```

Ramka danych zawiera następujące zmienne:


1. Type_of_Renewable_Energy: Numerical code representing the type of renewable energy source (1: Solar, 2: Wind, 3: Hydroelectric, 4: Geothermal, 5: Biomass, 6: Tidal, 7: Wave). 
2. Installed_Capacity_MW: Installed capacity in megawatts (MW). 
3. Energy_Production_MWh: Yearly energy production in megawatt-hours (MWh).
4. Energy_Consumption_MWh: Yearly energy consumption in megawatt-hours (MWh).
5. Energy_Storage_Capacity_MWh: Energy storage capacity in megawatt-hours (MWh).
6.  Storage_Efficiency_Percentage: Efficiency of energy storage systems in percentage.
7.  Grid_Integration_Level: Numerical code representing the level of grid integration (1: Fully Integrated, 2: Partially Integrated, 3: Minimal Integration, 4: Isolated Microgrid).
8.  Initial_Investment_USD: Initial investment costs in USD.
9.  Funding_Sources: Numerical code representing the funding source (1: Government, 2: Private, 3: Public-Private Partnership).
10.  Financial_Incentives_USD: Financial incentives in USD.
11.  GHG_Emission_Reduction_tCO2e: Reduction in greenhouse gas emissions in tons of CO2 equivalent (tCO2e).
12. Air_Pollution_Reduction_Index: Air pollution reduction index.
13. Jobs_Created: Number of jobs created.

Zależności w zbiorze danych:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.drop(columns=['Type_of_Renewable_Energy']).corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Macierz korelacji")
plt.show()
```

Zmienne są niezależne, nie ma między nimi korelacji.

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df, x='Funding_Sources', y='Energy_Production_MWh', estimator=sum, color = 'green')
plt.title('Produkcja energii')
plt.xlabel('Źródło finansowania')
plt.ylabel('Całkowita produkcja energii') 
plt.tight_layout()
labels = ['Government', 'Private', 'Public-Private Partnership']
ax.set_xticklabels(labels, rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df, x='Type_of_Renewable_Energy', y='Installed_Capacity_MW', estimator='mean' , color = 'green')
plt.title('Średnia pojemność')
plt.xlabel('Typ odnawialnego źródła energii')
plt.ylabel('Całkowita pojemność (MW)')
plt.tight_layout()
labels = ['Solar', 'Wind', 'Hydroelectric', 'Geothermal', 'Biomass', 'Tidal', 'Wave']
ax.set_xticklabels(labels, rotation=45)
plt.show()
```

Wysokości słupków są bardzo zbliżone niezależnie od tego, jak grupujemy.

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Type_of_Renewable_Energy', y='Energy_Production_MWh', data=df, color = 'green')
plt.title('Produkcja energii dla różnych typów odnawialnej energii')
plt.xlabel('Typ odnawialnej energii')
plt.ylabel('Produkcja energii (MWh)')

labels = ['Solar', 'Wind', 'Hydroelectric', 'Geothermal', 'Biomass', 'Tidal', 'Wave']
ax.set_xticklabels(labels, rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Type_of_Renewable_Energy', y='Energy_Storage_Capacity_MWh', data=df, color = 'green')
plt.title('Przechowanie energii dla różnych typów odnawialnej energii')
plt.xlabel('Typ odnawialnej energii')
plt.ylabel('Energia przechowywana (MWh)')

labels = ['Solar', 'Wind', 'Hydroelectric', 'Geothermal', 'Biomass', 'Tidal', 'Wave']
ax.set_xticklabels(labels, rotation=45)
plt.tight_layout()
plt.show()
```

Rozkłady zarówno produkcji energii dla różnych źródeł oraz pojemności przechowywania energii dla różnych źródeł praktycznie się niczym między sobą nie różnią.
