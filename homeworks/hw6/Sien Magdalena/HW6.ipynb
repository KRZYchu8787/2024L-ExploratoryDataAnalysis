```python
import matplotlib.pyplot as plt
import numpy as npA
import pandas as pd
import seaborn as sns
df=pd.read_csv("energy_dataset_.csv")
```

## Opis zbioru danych 

```python
print("Zbior danych zawiera informacje nt. odnawialnych systemow energii i ma ",df.shape[0]," wierszy oraz ",df.shape[1]," kolumn.\nWystepuja w nim zmienne o nastepujacych nazwach i typach:\n",df.dtypes,sep='')
```

## Analiza zmiennych

```python
print("Statystyki zmiennych z ramki danych:\n",df.describe())
```

## Analiza zmiennych w zależności od przynależności do grupy

```python
print("Statystyki zmiennych pogrupowane przez zmienna Type_of_Renewable_Energy:\n",df.groupby("Type_of_Renewable_Energy").describe(),"\nNumerowe wartosci odpowiadaja nastepujacym typom energii:\n1: Solar, 2: Wind, 3: Hydroelectric, 4: Geothermal, 5: Biomass, 6: Tidal, 7: Wave.\nMozna zauwazyc, ze statystyki sa zblizone, a niektore wrecz takie same, niezaleznie od typu energii odnawialnej. ",sep='')
```

## 2 wykresy analizujące rozkład zmiennych w zbiorze danych

```python
sns.boxplot(data=df.loc[df.GHG_Emission_Reduction_tCO2e<500],x='Grid_Integration_Level', y='GHG_Emission_Reduction_tCO2e')
print("Numerycznym wartosciom zmiennej Grid_Integration_Level odpowiadaja nastepujace wartosci:\n1: Fully Integrated, 2: Partially Integrated, 3: Minimal Integration, 4: Isolated Microgrid.\n")
plt.title("Rozklad zmiennej GHG_Emission_Reduction_tCO2e z podzialem na Grid_Integration_Level z limitem 500 tCO2e")
plt.show()
print("\nWNIOSKI:\nDla malych wartosci zmiennej GHG_Emission_Reduction_tCO2e (<500 tCO2e) można zauwazyc, ze srednia wartosc najwyzsza jest dla typu\n4 - Isolated Microgrid, a najnizsza dla typu 1 - Fully Integrated. Srednie te roznia sie o okolo 100 tCO2e.\nTo swiadczy o tym, iz udalo sie najepiej zredukowac emisje gazow, stosujac izolowana mikrosiec.")
```

```python
sns.violinplot(x='Funding_Sources',
               y='Jobs_Created', 
               data=df.loc[df.Jobs_Created])
plt.title('Rozklad zmiennej Jobs_Created z podzialem na Funding_Sources')
print("Numerycznym wartosciom zmiennej Funding_Sources odpowiadaja nastepujace wartosci:\n1: Government, 2: Private, 3: Public-Private Partnership.\n")
plt.show()
print("\nWNIOSKI:\nNiezależnie od zrodla finansowania zmienna dotyczaca powstalych miejsc pracy ma zbizony rozklad i podobna srednia wartosc w okolicy 2500 miejsc pracy.\nMozna wysnuc wniosek, iz zrodlo finansowania nie wplywa na ilosc stworzonych miejsc pracy.")
```

## 2 wykresy wskazujące zależności w zbiorze danych z wnioskami

```python
sns.scatterplot(x='GHG_Emission_Reduction_tCO2e',y='Air_Pollution_Reduction_Index',data=df.loc[(df.GHG_Emission_Reduction_tCO2e<500)])
plt.title('Zaleznosc zmiennej Air_Pollution_Reduction_Index od GHG_Emission_Reduction_tCO2e')
plt.show()
print("WNIOSKI:\nDla malych wartosci zmiennej GHG_Emission_Reduction_tCO2e (<500 tCO2e) nie widac zaleznosci miedzy GHG_Emission_Reduction_tCO2e a\nAir_Pollution_Reduction_Index. Mozna stwierdzic, ze zmienne nie sa ze soba skorelowane.")
```

```python
sns.scatterplot(x='Energy_Production_MWh',y='Energy_Consumption_MWh',data=df.loc[df.Energy_Consumption_MWh<5000])
plt.title("Zaleznosc zmiennej Energy_Consumption_MWh od Energy_Production_MWh")
plt.show()
print("\nWNIOSKI:\nWbrew temu, co mozna bylo sie spodziewac, dla ograniczonych wartosci zmiennej Energy_Consumption_MWh (<5000) nie widac zwiazku miedzy zmiennymi\nEnergy_Consumption_MWh i Energy_Production_MWh. Mozna stwierdzic, iz nie istnieje miedzy nimi korelacja.")
```
