## Rozkład zmiennych i wnioski

```python
import seaborn as sns
sns.set(rc = {'figure.figsize': (12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows = 10
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
```

```python
plot=sns.violinplot(y = 'Ladder score', data=df)
plot.set_title("Distribiuton of ladder score in 2024 (on scale from 0 to 10)");
plot.set_ylabel("Ladder score");
```

<div style="text-align: justify;">
Z powyższego wykresu widzimy, że mediana poziomu szczęścia na świecie osiąga wartość ok. 5.8, zaś dla wartości ok. 6.1    istnieje najwięcej krajów osiągający ten poziom szczęścia. Zatem najwięcej krajów osiąga "trochę powyżej przeciętny" poziom szczęścia. W okolicach warości 7.5 następuje gwałtowny spadek liczby krajów, co ocznacza, że krajów o bardzo wysokim poziomie szczęścia jest dość mało. Z drugiej strony dla wartości poniżej 6 także następuję spadek liczności krajów o danym wyniku, jednak jest on znacznie wolniejszy niż dla wartości powyżej 6 co powoduje, że rozkład naszej zmiennej jest lewostronnie asymptotyczny z długim ogonem. W okolicach 2.5 gwałtwonie spada liczba krajów o tym poziomie szczęścia, co implikuje, że jest niewiele krajów o bardzo niskim poczuciu szczęścia.
</div>

```python
plot2=sns.violinplot(y="Log GDP per capita",data=df);
plot2.set_title("Distribiuton of GDP per capita");
plot2.set_ylabel("Log GDP per capita");
```

<div style="text-align: justify;">
Ponownie jak w poprzednim przypadku, mediana naszego zbioru (ok. 1.43) znajduje się minimalnie poniżej najczęściej występujących wartości (1.5-1.8). Potem następuje gwałtowny spadek w okolicah 2.0. Możemy wysnuć z tego wniosek, że najwięcej krajów ma PKB per capita na poziomie "średnim i średnim wyższym", ale za to jest bardzo mało krajów, które byłyby bardzo bogate. Dla wartości poniżej medinay następuję powolny spadek, co pokazuje nam długi lewostronny ogon. Jednak dla wartości ok. 0.75 spadek mocno przyśpiesza, co ponownie umożliwia wysnucie wniosku, iż liczba bardzo biednych krajów jest niewielka. Wybrzuszenie w 0.0 jest nieistotne, gdyż jest to oznaczenie braku danych.
</div>

## Zależności  między danymi i wnioski

```python
plot3=sns.scatterplot(data=df, x='Ladder score', y='Healthy life expectancy', hue='Regional indicator')
sns.regplot(data=df, x='Ladder score', y='Healthy life expectancy', ax=plot3, scatter=False, color='red');
plot3.set_title("Relationship between healthy life expectancy and ladder score");
```

<div style="text-align: justify;">
Z wykresu możemy odczytać, że istnieje znaczna dodatnia korelacja pomiędzy poziomem szczęścia, a przewidywaną długością życia w zdrowiu, co dokładnie obrazuje nam czerwona linia regresji (współczynnik korelacji na poziomie 0.77). Widzimy na przykład, że kraje regionu Afryki Subsaharyjskiej charakteryzują się niskim poziomem szczęścia, ale też krótką długością życia w zdrowiu. Na drugim biegunie znajduje się cała Europa, która posiada wysoki poziom szczęścia i długą długość życia w zdrowiu. Jedynym regionem, który istotnie wyłamuje się tej regule jest Bliski Wschód i Afryka Północna, która pomimo dość długiej oczekiwanej długości życia w zdrowie nie charakteryzuje się wysokim poczuciem szczęśćia.
</div>

```python
plot4=sns.scatterplot(data=df, x='Log GDP per capita', y='Social support', hue='Regional indicator')
sns.regplot(data=df, x='Log GDP per capita', y='Social support', ax=plot4, scatter=False, color='red');
plot4.set_title("Relationship between social support and GDP per capita");
```

<div style="text-align: justify;">
Kolejną dodatnią korelacją między dwoma z zmiennymi możemy zauważyć, jeśli porównamy ze sobą PKB per capita i poczucie wsparcia przez osoby bliskie. Widać tutaj jak bardzo kwestie finansowe całego państwa mogą wpływać na realcje pojedynczych jednostek.  Taki wniosek możemy wysnuć po obserwacji linii regresji z dość wąski przedziałem ufności poza obszarem o niskim PKB, co wynika jednak z braku dużej liczby obserwacji. Większość obserwacji dość dobrze spełnia powyższą zależność. Jedyne kraje, które w znaczny spsób wyłamują się powyższej tendencji to kilka krajów Afryki i Azji Południowej.
</div
