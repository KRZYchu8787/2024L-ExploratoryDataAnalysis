````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize': (12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows = 10
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
````

## Opis danych
Ramka ```laptops.csv``` zawiera dane o laptopach wystawionych na sprzedaż w platformie eBay. Ma niecałe 3000 wierszy i 10 kolumn, które zawierają informacje o modelu, parametrach, stanie i cenie laptopa.

````python
df = pd.read_csv('laptops.csv')
df.head()
````

````python
df.describe()
````

Wykres poniżej potwierdza, że laptopy mają zazwyczaj pewne ustalone przekątne ekranu; rzadko trafia się laptop, który ma ekran o przekątnej pomiędzy 14" a 15".

````python
sns.displot(data = df, x = "Screen_Size")
````

````python
sns.catplot(y = "Brand",
           x = "Price", 
           data = df,
           kind = "box")
````

Z powyższego wykresu wynika, że laptopy firmy MSI i Gigabyte są drogie. Dodatkowo, laptopy marki Panasonic mają dość duży rozstrzał w cenach; występują outliery. Najbardziej zróżnicowane są ceny laptopów Gigabyte, choć największe różnice między najmniejszą a największą ceną występują w pierwszych 5 markach.

````python
sns.scatterplot(data = df, x = "Screen_Size", y = "Price")
````

Jak widać, największy rozstrzał cen ma miejsce dla laptopów o ekranach 14 calowych. Wynika to z faktu, że jest to najpopularniejszy ich rozmiar. Dodatkowo, w przypadku dużych wymiarów można zauważyć, że ceny są wyższe niż w przypadku mniejszych przekątnych.

````python
sns.scatterplot(data = df.sample(n=400), y = "Resolution", x = "Brand", hue = "Processor")
````
